# Subdividing into seasonal data

The data contains information on each League of Legends regional competition as well as the World Championship.This specific dataset was chosen as it is a collation of all the matches of 2021, which has an in-game meta closely tied to the meta of 2022.

This notebook is for exploratory analysis. League is a complex game where only multiple views on matches can lead to solid predictions, but even that may not be enough. I will use this exploratory analysis in combination with my own knowledge on the game (played for many years and reached the top 1% of the playerbase). It can be used for eSports betting, as a good insight into each region is given. 

Credit to Tim Sevenhuysen from Oracle's Elixir for providing the dataset, which can be found [here](https://oracleselixir.com/tools/downloads).

## Some thoughts


* Makes sense that the As a lot of unique picks since they play in a 'best-of-3' format, so the players can take more risk with the safety net of extra games available to them. In contrast, LCS and LEC players are far less experimental, perhaps due to the 'best-of-1' format where they can get punished without a chance to adapt in future games. 
  
  
* There is a high level of variety coming from LCK top laners, relative to the LPL, despite a similar number of games played. LCK truly live up to their reputation as the home of the top lane, having produced some of the most prodigiously talented top laners throughout all of league history, and as such it only logically follows that there would be a higher degree of creativity coming out from the top lane. Generally that level of willingness to experiment with unique champion picks seems to be present across all roles, as they seem to be willing to pick through a wider range of champions than other regions are. Keep in mind the context that this is all dealing with champion picks in terms of volume and does not necessarily deal with the qualitative variables that can factor into what is good experimentation vs bad experimentation, ie matchup, style of play as a team, etc.  
 
 
* Despite their large number of games, the LPL is not the most explorative. In a roundabout way, this can make sense. The LPL is known to play by their own rules; they understand the aggressive playstyle suits them, and stick to it. The LPL ladder demands capitalising on every mistake given the millions of people playing on it (far greater than other regions) and so it is only natural for this to be the dominant playstyle in pro-play. Any deviation from this would require a significant time investment that coaches are probably unwilling to make due a possible regress in competetive standing.  


* With regards to LCS vs LEC, 2 leagues with similar formats, one interesting thing that stood out to me was the increased variety in mid lane in the LCS. This is very surprising as it suggests that the LCS was still trying to find its meta-playstyle in the summer, whereas perhaps LEC had already found it. Although, it could also go against conventional thinking and suggest that the LCS was utilising the carry potential of mid as much as other regions, whereas the LEC were trying to play a more balanced game.

There were barely any LPL imports into these regions relative to the number of LCK imports, and so it could mean they don't think a controlled scaling style would work against LPL's aggression in the world championship. 


* Surprised to see a Twisted Fate being picked at worlds, having only been prevelant in the LEC beforehand. This could be due to a meta change whereby LEC teams were winning with the pick, forcing the hands of their enemies to then adapt and copy the playstyle.

I looked it up and it turns out she was nerfed right before worlds, in [Patch 11.9](https://www.leagueoflegends.com/en-gb/news/game-updates/patch-11-9-notes/). This affected her clear speed and meant that she could not match other jungles to the first crab. In a meta where it is crucial to get the first crab and dominate the enemy jungler, this makes a lot of sense. 


* Talon seemingly came out of nowhere in the world championship, despite having not been buffed in the worlds patch. Most likely just an insane scrim pick that somebody found. Important to note that it would have been picked a lot more if it wasn't banned as much as it was. In the games where it was played, it looked very strong. 

# Game Length

```python
ylabels= [lpl_summer_data['gamelength'].mean(), lcs_summer_data['gamelength'].mean(), lec_summer_data['gamelength'].mean(), lck_summer_data['gamelength'].mean()]
xlabels = ['LPL', 'LCS', 'LEC', 'LCK']
plt.bar(xlabels, ylabels);
plt.title("Summer Average Game Time")
x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2, 1500, 2100))
plt.xlabel('League');
plt.ylabel('Game time in seconds');
```

```python
ylabels= [chinese_data['gamelength'].mean(), na_data['gamelength'].mean(), eu_data['gamelength'].mean(), kr_data['gamelength'].mean(), other_data['gamelength'].mean()]
xlabels = ['LPL', 'LCS', 'LEC', 'LCK', 'OTHERS']
plt.bar(xlabels, ylabels);
x1,x2,y1,y2 = plt.axis()
plt.axis((x1,x2, 1500, 2300))
plt.title("Average Game Time At Worlds")
plt.xlabel('League');
plt.ylabel('Game time in seconds');
```

## Some thoughts

* Makes sense for the LCK to have the longest average game time since they play the most controlled and scaling playstyle, waiting for their carries to be strong enough to win.

* On the surface, it is odd to see the LCS have long game times as they pick aggressive early game champions similar to those of the LPL and LEC. However, anybody who knows about the infamous 'NArams' that occur can see why games are prolonged. The teams simply don't utilise their champions correctly and so 'messy' game states often occur where teams go back and forth in fights depending on who played better mechanically, irrespective of team compositions. 

* Interesting that the LEC had the longest games at worlds. Despite not winning many of their games compared to the eastern teams, they were able to prolong the games and prevent quick stomps from happening. This is a testament to the mental strength EU teams show every year. In comparison, the LCS teams lost a lot more and had shorter game times. From watching the games, it is obvious why. Most games were heavily one-sided against their eastern opponenets and unlike the LEC teams, they could not prolong their inevitable defeat.

# Some thoughts

* The distributions are very similar in terms of structure, and the small variations can be attributed to differences in the number of games played. 


* When you compare the LEC graph to the LCS graph there is a clear disparity in that the LCS tail is even skinnier than the LEC one. For some reason the style of play itself within the LCS region is far less conducive to high gold differences. One idea that could be posited to account for this difference could be the general passivity and the focus on team fights in the NA LCS region. While everyone does remark that the style of play in the LCS usually does come down to team fights, as opposed to other regions like LEC where general meta strategies like side laning also thrive, there actually are not as many skirmishes or team fights as there is in a region like LPL. In the case of LCS the region exhibits a tendency towards team fights in conjunction with a general level of passivity that, at least in my eyes, can inhibit truly large levels of gold leads from being generated.
